## Comparing Coastguard Cleaners

```python
#Import Modules
import psrchive
import matplotlib.pyplot as plt
import numpy as np
```

### Only CoastGuard

```python
#Each cleaner alone
bw, hb, rstd, surg = 19.785, 41.386, 19.785, 402.519

#Each cleaner with surgical
bw_surg, hb_surg, rstd_surg, bw_hb_surg, bw_rstd_surg, hb_rstd_surg, all_cleaners = 432.599, 397.118, 402.519, 426.797, 432.5999, 397.561, 426.730

cleaners = [bw, hb, rstd, surg, bw_surg, hb_surg, rstd_surg, bw_hb_surg, bw_rstd_surg, hb_rstd_surg, all_cleaners]
labels = ['Bandwagon', 'Hotbins', 'Rcvrstd', 'Surgical', 'Bandwagon+Surgical', 'Hotbins+Surgical', 'Rcvrstd+Surgical', 'Bandwagon+Rcvrstd+Surgical','Bandwagon+Rcvrstd+Surgical', 'Hotbins+Rcvrstd+Surgical',  'All_Cleaners']
x = range(0,len(cleaners))
```

```python
clrs = ['tab:blue' if (x < max(cleaners)) else 'tab:green' for x in cleaners ]
plt.bar(x, cleaners, color=clrs)
plt.xticks(x, labels, color='black', rotation=45, horizontalalignment='right')
plt.ylabel('SNR')
plt.xlabel('Cleaner(s) Used')
plt.title('Only Coastguard')
plt.show()
plt.close()
```

### CLFD + CoastGuard

```python
clfd_all, clfd_Bandwagon_Rcvrstd_Surgical, clfd_Bandwagon_Surgical, clfd_Bandwagon_Rcvrstd_Surgical_Manual = 430.631, 432.635, 432.635, 425.9
clfd_cleaners = [clfd_all, clfd_Bandwagon_Rcvrstd_Surgical, clfd_Bandwagon_Surgical, clfd_Bandwagon_Rcvrstd_Surgical_Manual]
clfd_labels = ['All_Cleaners', 'Bandwagon+Rcvrstd+Surgical', 'Bandwagon+Surgical', 'Bandwagon+Rcvrstd+Surgical+Manual']
x2 = range(0,len(clfd_cleaners))
```

```python
clrs2 = ['tab:blue' if (x < max(clfd_cleaners)) else 'tab:green' for x in clfd_cleaners ]
plt.bar(x2, clfd_cleaners, color=clrs2)
plt.xticks(x2, clfd_labels, color='black', rotation=45, horizontalalignment='right')
plt.ylabel('SNR')
plt.xlabel('Cleaner(s) Used')
plt.title('CLFD + Coastguard')
plt.ylim([420, 440])
plt.show()
plt.close()
```
